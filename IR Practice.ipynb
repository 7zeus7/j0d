# Write a program to demonstrate bitwise operation.

```python
list1=[2,4,6,90]
list2=[3,4,6,90]
def postingList(l1,l2):
    p1=0
    p2=0
    ans=[]
    while p1!=len(l1) and p2!=len(l2):
        if l1[p1]==l2[p2]:
            ans.append(l1[p1])
            p1+=1
            p2+=1
        elif l1[p1]>l2[p2]:
            p2+=1
        else :
            p1+=1
    return ans
postingList(list1,list2)
```

# Implement Dynamic programming algorithm for computing the edit distance between strings s1 and s2. 

```python
def editDistance(s1,s2,n,m):
    if n==0:
        return m
    if m==0:
        return n
    if(s1[n-1]==s2[m-1]):
        return editDistance(s1,s2,n-1,m-1)
    return 1+min(editDistance(s1,s2,n,m-1),editDistance(s1,s2,n-1,m-1))
s1='sanjn'
s2='sanjana'
print(editDistance(s1,s2,len(s1),len(s2)))
```

# Write a program to implement simple web crawler.

```python
import requests
from parsel import Selector
import time

start_time=time.time()
response=requests.get("http://recurship.com/")
selector=Selector(response.text)
href=selector.xpath("//a/@href").getall()
image=selector.xpath("//img/@src").getall()

end_time=time.time()

print('---------------link--------------')
print(href)
print('---------------image--------------')
print(image)
print("time taken ",end_time-start_time)
```

# Write a program to parse XML text and save it to CSV

```python
import csv
import requests
import xml.etree.ElementTree as ET
import pandas as pd

def loadRSS():
    url = "https://timesofindia.indiatimes.com/rssfeeds/1081479906.cms"
    resp = requests.get(url)
    with open('topnewsfeed.xml','wb') as f:
        f.write(resp.content)
        
def parseXML(xmlfile):
    tree = ET.parse(xmlfile)
    root = tree.getroot()
    newsitems = []
    for item in root.findall('./channel/item'):
        news = {}
        for child in item[:5]:
            news[child.tag] = child.text.encode('utf8')
        newsitems.append(news)
    return newsitems

def savetoCSV(newsitems,filename):
    df=pd.DataFrame(newsitems)
    df.to_csv(filename)

def main():
    loadRSS()
    newsitems = parseXML('topnewsfeed.xml')
    savetoCSV(newsitems,"topnews.csv")

if __name__ == "__main__":
    main()
```
